```python
import altair as alt
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv('iris.csv')
iris.info()
iris.head(5)
iris.describe()
```

```python
alt.Chart(iris).mark_point(filled=True,size=50).encode(
    alt.X('SepalLengthCm:Q', scale=alt.Scale(domain= [4,8]), axis=alt.Axis(title='Sepal Length')),
    alt.Y('SepalWidthCm:Q', scale=alt.Scale(domain= [1.5,5]), axis=alt.Axis(title='Sepal Width')),
    color='Species:N',
    tooltip=['Species', 'SepalLengthCm', 'SepalWidthCm']
).interactive()
```

```python
alt.Chart(iris).mark_bar().encode(
    x='PetalWidthCm:Q',
    y='count(PetalLengthCm):Q',
    color='Species:N',
    facet=alt.Facet('Species:N', title=None)
).properties(width=200, height=150)
```
